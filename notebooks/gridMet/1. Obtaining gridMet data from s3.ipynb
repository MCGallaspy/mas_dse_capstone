# About this notebook

This notebook demonstrates how to obtain gridMet data from Amazon S3.
Please install the `s3fs` and `boto3` python libraries.
If you have authentication errors, see this page for setting up your AWS credentials with `boto3`: https://boto3.amazonaws.com/v1/documentation/api/latest/guide/quickstart.html#configuration

```python
import boto3
import pandas as pd
```

```python
s3 = boto3.client("s3")
all_objects = s3.list_objects(Bucket='dse-cohort5-group5')
all_objects
```

The file we want is at `s3://dse-cohort5-group5/wildfire_capstone/gridMet.parquet.gz`

```python
%%time
df = pd.read_parquet("s3://dse-cohort5-group5/wildfire_capstone/gridMet.parquet.gz")
```

```python
print(df.shape)
df.head()
```

```python
df.to_parquet("gridMet.parquet.gz", compression="gzip")
```
