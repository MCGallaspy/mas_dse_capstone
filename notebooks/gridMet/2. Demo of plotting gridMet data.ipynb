```python
import pandas as pd
import geopandas as gpd
import numpy as np
import contextily as ctx
from matplotlib import pyplot as plt
```

```python
%%time
df = pd.read_parquet('gridMet.parquet.gz').reset_index()
```

```python
df.head()
```

```python
df.date.describe()
```

```python
%%time
# Use a mask of the original dataframe to cut down on computation
mask = df.date >= pd.to_datetime('2019', format="%Y")
df = gpd.GeoDataFrame(df[mask], geometry=gpd.points_from_xy(df[mask].longitude, df[mask].latitude))
df.head()
```

```python
df.crs = {'init' :'epsg:4326'}
df = df.to_crs(epsg=3857)
```

```python
from mpl_toolkits.axes_grid1 import make_axes_locatable
f, ax = plt.subplots(1, figsize=(12, 12))
divider = make_axes_locatable(ax)
cax = divider.append_axes("right", size="5%", pad=0.05)
date = pd.to_datetime('2019-01-01', format="%Y-%m-%d")
date_str = date.strftime("%Y-%m-%d")
mask = df.date == date
cbar_tick_delta = df[mask]["relative_humidity_%"].max() - df[mask]["relative_humidity_%"].min()
cbar_ticks=[df[mask]["relative_humidity_%"].min() + cbar_tick_delta * 0.2 * i for i in range(0, 6)]
df[mask].plot(
    ax=ax,
    column='relative_humidity_%',
    cmap='plasma',
    legend=True,
    legend_kwds={"cax": cax, "ticks": cbar_ticks},
    alpha=.8)
ax.set_xticks([])
ax.set_yticks([])
ax.set_xticklabels([])
ax.set_yticklabels([])
ax.set_title("Relative humidity (%) for San Diego county on {}".format(date_str))
ctx.add_basemap(ax)

f, ax = plt.subplots(1, figsize=(4, 4))
ax.set_title("Histogram of relative humidity\n"
             "for locations in San Diego county\n"
             "on {}".format(date_str))
ax.set_xlabel("Relative humidity (%)")
ax.set_ylabel("Count")
df[mask]['relative_humidity_%'].hist(ax=ax, bins=100);
```
